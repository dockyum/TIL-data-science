# 타이타닉 데이터 결측치 채우기

```python
titanic_df = sns.load_dataset("titanic")
```

```python
msno.matrix(titanic_df)
```

```python
# 80% 이상의 컬럼 데이터만 남기고 
# 범주형 데이터는 최빈값
# 수치형 데이터는 평균이나 중앙값으로 채우기
```

```python
# 80% 이상의 컬럼 데이터만 남기고 
titanic_df = titanic_df.dropna(thresh=int(len(titanic_df)*0.8), axis=1)
```

```python
msno.matrix(titanic_df)
```

```python
titanic_df.columns
```

```python
titanic_df.info()
```

```python
titanic_df.head(2)
```

```python
## categorical data => most frequent
columns = ["embarked", "embark_town"]
imputer_mf = SimpleImputer(strategy="most_frequent")
```

```python
titanic_df['embarked'].head()
## Series
```

```python
titanic_df[['embarked']].head()
## DataFrame
```

```python
for column in columns:
    titanic_df[column] = imputer_mf.fit_transform(titanic_df[[column]])
```

```python
msno.matrix(titanic_df)
```

```python
## continuous data => median (중앙값) or mean
```

```python
sns.displot(titanic_df.age)
```

```python
imputer_median = SimpleImputer(strategy="median")
```

```python
titanic_df["age"] = imputer_median.fit_transform(titanic_df[['age']])
```

```python
msno.matrix(titanic_df)
```

## 2. 범주형 데이터
- categorical data
- 더미 변수화 : 단순 분류 데이터. 순서가 없어야 할 경우
    - 예를 들어 embarked의 위치에 따라 사망확률이 다르다면, 점수를 매겨도 됨
    - 원핫 인코딩
    - 전부 해보고 결과가 좋게 나오는 것으로 하면 됨
 

```python
titanic_df.columns
```

```python
df = titanic_df[['survived','pclass', 'sex', 'age','embarked']]
df.head()
```

```python
pd.get_dummies(df[['sex']]).head()
```

```python
sex_df = pd.get_dummies(df["sex"])
embarked_df = pd.get_dummies(df["embarked"])
```

```python
df = pd.concat([df,sex_df, embarked_df], axis=1)
df
```

```python
df = df.drop(columns=["sex","embarked"])
df
```
